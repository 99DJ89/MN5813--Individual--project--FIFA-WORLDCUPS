The Olympic Games immediately bring to mind powerful images of grandeur, excellence, internationalism, and history. For more than a millennium, from approximately 776 BC to about 400 AD, the ancient Olympics served as a powerful cultural institution in Greece. Within this immense history, truly unique accomplishments occurred, such as the astounding feat achieved by the runner Polites of Caria in 69 AD. Polites accomplished the nearly incredible challenge of winning both the shortest sprint, the stade (about 200 meters), and the long distance race on the same day. This demonstration of versatility was considered a "great marvel" in antiquity by Pausanias, and no athlete in the ancient or modern eras has ever equaled this particular double victory. Modern coaches believe long and short distances require wholly different training, muscle fibers, and techniques.
It is appropriate to compare ancient and modern competitors, as our contemporary athletic system is the only one in world history that is substantially similar to the Greeks’ system, having borrowed heavily from it. However, the terminology itself requires clarification. The popular phrase “Olympic Games” is considered a bad mistranslation. The correct Greek term, Olympiakoi agones, does not translate to "games" or "play," but rather to “struggles” or “contests,” which is the root of the English word “agony”. Greek athletics were focused on individual contests and were not considered recreational "sport" or diversionary festivals.
Furthermore, historical research into the ancient games reveals that the concept of amateurism was not practiced by the ancient Greeks, despite the insistence of figures like Avery Brundage in the modern era. The modern revival itself—specifically track and field—arose from a conscious imitation of ancient Greek practices, promoted by early revivals in Britain and modern Greece beginning as early as 1859. This history aims to provide an accessible and comprehensive look at these ancient competitions and their influence on the world.

## Data management

### Data cleaning

#### Data cleaning is essential before analysis. We'll identify and handle:
-Missing values
-Data type issues

```python
# TODO: Handle missing values
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(1)

missing_summary = pd.DataFrame({
    "Missing_Count": missing,
    "Missing_Percent": missing_pct
})

display ("Missing values summary:")
display (missing_summary[missing_summary['Missing_Count'] > 0].sort_values("Missing_Count", ascending=False))
```

## Analysis of missing values:

- Medal (85.3%): This is expected - most employees don't receive awards, just like most athletes don't win medals
- Weight_kg (22.8%): Demographic data may be incomplete
- Weight_cm (21.8%): Similar to weight
- Age (3.5%): Personal information may not always be recorded
### Decision: Keep missing values as-is for demographic columns (age, height, weight) since:
1. The percentage is relatively low
2. Imputing (filling) could introduce bias
3. Analysis functions like `.mean()` handle NaN automatically

For **award**, NaN represents "no award" which is meaningful information.

For Olympics: You'll see similar patterns - most athletes have no medals, and some demographic data is missing.

### Data wrangling

```python
# TODO: Create at least one new column
# Create new features: BMI (Body Mass Index)
df['BMI'] = df['Weight'] / (df['Height']/100)**2
```

```python
# TODO: Combine at least two columns
# Combine 'Games' and 'Year' for analysis
df['Games_Year'] = df['Games'].astype(str) + ' (' + df['Year'].astype(str) + ')'
```

```python
# TODO: Extract data from multiple columns to create a new column
# Country & Medal Status
df['Has_Medal'] = df['Medal'].notnull().astype(int)
```

```python
# TODO: Anything else?
# Categorizing Age Groups (for demographic analysis)
bins = [0, 18, 30, 45, 100]
labels = ['Youth','Adult','Mid-age','Senior']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
# Normalize Sport Names (strip spaces, lowercase)
df['Sport_Norm'] = df['Sport'].str.strip().str.lower()
```

```python
# Create a unique identifier for each athlete event
df['AthleteEventID'] = df['ID'].astype(str) + '_' + df['Event'].astype(str) + '_' + df['Year'].astype(str)
```

```python
# Preview wrangled columns
display (df[['BMI', 'Games_Year', 'Has_Medal', 'Age_Group', 'Sport_Norm', 'AthleteEventID']].head(10))
```
